```python
col=['Total_MVA','Po_GFM_MVA','Po_SG_MVA','Po_GFL_MVA','Fre_SG_Hz','Fre_GFM_Hz', 'Fre_GFL_Hz', 'Vo_SG_PU','Vo_GFM_PU', 'Vo_GFL_PU']
col_c=['countT','countGM','countS','countGL','c5','c6','c7','c8','c9','c10']
dfr=0
for i in range(10):
    dfr = pd.DataFrame(df[col[i]].value_counts())
    dfr.reset_index(inplace=True)
    dfr.columns = [col[i],col_c[i]]
    dfr = df.merge(dfr,on=col[i])
    df=dfr
    print(df)
```

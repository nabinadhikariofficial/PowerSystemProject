```python
col=['Total_MVA','Po_GFM_MVA','Po_SG_MVA','Po_GFL_MVA']
col_c=['countT','countGM','countS','countGL']
dfr=0
for i in range(4):
    dfr = pd.DataFrame(df[col[i]].value_counts())
    dfr.reset_index(inplace=True)
    dfr.columns = [col[i],col_c[i]]
    dfr = df.merge(dfr,on=col[i])
    df=dfr
    print(df)
```
